```python
import numpy as np
import pandas as pd
```

```python
cardio = pd.read_csv("../data/raw/cardio.csv")
cardio = cardio.iloc[:, 1:]
cardio

# cardio.to_csv("../data/raw/cardio.csv", sep=",", index=False)
```

```python
cardio.columns.tolist()
```

```python
target_col = "cardio"
```

```python
# Identify numerical columns (excluding the target column)
num_col_idx = [i for i, col in enumerate(cardio.columns) if pd.api.types.is_numeric_dtype(cardio[col]) and col != target_col]

# Identify categorical columns (excluding the target column)
cat_col_idx = [i for i, col in enumerate(cardio.columns) if pd.api.types.is_object_dtype(cardio[col]) and col != target_col]

# Identify target column index
target_col_idx = [cardio.columns.get_loc(target_col)]
```
